## Problem 2

1. Write a recursive function `factorial()` that takes in a natural number `n` and computes $n!=n(n-1)(n-2)\cdots3\cdot2\cdot 1$. 
2. Use `%timeit` to determine how fast your function calculates `factorial(150)`. 
3. Use `%timeit` to calculate how fast `math.factorial()` calculates the same quantity.
4. Were there any differences between the two methods?

```python
def factorial_helper(n, total):
    if (n == 0):
        return total
    return factorial_helper(n-1, total * n)

def factorial(n):
    return factorial_helper(n, 1)

%timeit factorial(150)
%timeit math.factorial(150)

# the difference was mine was much slower. I'm pretty sure this is because math implements it in C since Python is so slow.
```

## Problem 4

For each of the following functions, use `sympy` calculate the first derivative, the second derivative, and the integral (all with respect to $x$):

1. $f(x)=x$
2. $f(x)=6x^{-1/2}$
3. $f(x)=\pi\sqrt{x}+x^5+7$
4. $f(x)=x(2x-5)(3+\tan(x))$

```python
import sympy as sp

x = sp.symbols('x')

functions = [
    x,
    6*x**(-sp.Rational(1, 2)),
    sp.pi*sp.sqrt(x) + x**5 + 7,
    x*(2*x - 5)*(3 + sp.tan(x))
]

for i, f in enumerate(functions, start=1):
    print("function ",i)
    print("f(x) =", f)
    print("First derivative:", sp.diff(f, x))
    print("Second derivative:", sp.diff(f, x, 2))
    print("Integral:", sp.integrate(f, x))
    print("-" * 10)
```

## Problem 5

From yesterday, we calculated <i>numerically</i> what the derivative of $\sin(x)$ is. 
1. Using sympy, calculate derivative at 0.
2. Using sympy, evaluate the limit definition of the derivative for $f(x)=\sin(x)$:
$$f'(x)=\lim_{h\to0}\frac{f(x+h)-f(x)}{h}
$$

```python
import sympy as sp

x, h = sp.symbols('x h')

f = sp.sin(x)

deriv_at_0 = sp.diff(f, x).subs(x, 0)
print("Derivative at x=0: ", deriv_at_0)

lim_def = sp.limit((sp.sin(x + h) - sp.sin(x)) / h, h, 0)
print("Limit definition: ", lim_def)
```
