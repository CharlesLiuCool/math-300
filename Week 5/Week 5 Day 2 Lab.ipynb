### Problem 1
Suppose $f(x)=x^2+6x+10$

1. Plot the function to visualize the minimum
2. Find the minimum using `scipy`

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

def f(x):
    return x**2 + 6*x + 10

x = np.linspace(-8, 4, 400)
y = f(x)

plt.figure(figsize=(6,4))
plt.plot(x, y, label=r"$f(x)=x^2+6x+10$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Plot of f(x)")
plt.grid(True)
plt.legend()

result = minimize(f, x0=0)

print("Minimum x:", result.x[0])
print("Minimum value:", result.fun)

plt.plot(result.x[0], result.fun, 'ro', markersize=8, label='Minimum')
plt.show()
````


## Problem 2

Find the maximum of $50x-x^2$.

````python
def f(x):
    return 50*x - x**2

def g(x):
    return -1 * f(x)

x = np.linspace(0, 50, 400)
y = f(x)

plt.figure(figsize=(6,4))
plt.plot(x, y, label=r"$f(x)=50x-x^2$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Plot of f(x)")
plt.grid(True)
plt.legend()

result = minimize(g, x0=0)

print("Maximize x:", result.x[0])
print("Maximize value:", -1 * result.fun)

plt.plot(result.x[0], -1 * result.fun, 'ro', markersize=8, label='Maximum')
plt.show()
````

## Problem 3

Minimize

```python
np.sin(x)+x**2
```

using different starting values.

Do you obtain the same solution? Compute the derivative using `sympy`. When is this derivative zero?

````python
import numpy as np
from scipy.optimize import minimize

def f(x):
    return np.sin(x) + x**2

starts = [-10, -5, 0, 5, 10]

for start in starts:
    result = minimize(f, x0=start)
    print(f"Start = {start}")
    print("Minimum x:", result.x[0])
    print("Minimum value:", result.fun)
    
# I always obtain the same solution

import sympy as sp

x = sp.Symbol('x')

f = sp.sin(x) + x**2
df = sp.diff(f, x)

print(df)
solution = sp.nsolve(df, -0.5)
print(solution)

# This solution matches up
````
