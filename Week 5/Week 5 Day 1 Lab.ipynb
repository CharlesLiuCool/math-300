### Problem 3
Make a 2 by 2 invertible matrix, A.
1. Compute its eigenvectors and plot them using matplotlib
2. Compute A times the eigenvectors and plot them on the same plot.
3. What do you notice?

```python
import matplotlib.pyplot as plt

A = np.array([[1,2],[3,1]])

eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:")
print(eigenvalues)

print("\nEigenvectors (columns):")
print(eigenvectors)

plt.figure(figsize=(6,6))
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

eigen_colors = ['blue', 'green']
transformed_colors = ['red', 'orange']

for i in range(2):
    v = eigenvectors[:, i]
    Av = A @ v

    plt.quiver(0, 0, v[0], v[1],
               angles='xy', scale_units='xy', scale=1,
               color=eigen_colors[i],
               label=f"Eigenvector {i+1}")

    plt.quiver(0, 0, Av[0], Av[1],
               angles='xy', scale_units='xy', scale=1,
               color=transformed_colors[i],
               label=f"A·v{i+1}")

plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.gca().set_aspect('equal')
plt.grid(True)
plt.legend()
plt.title("Eigenvectors and Their Images Under A")
```
