### Problem 1
Suppose 

$$\frac{dy}{dt}=1.2y.$$

instead of .5 from lecture. Solve again.

How does the graph change?



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def growth(t, y):
    return 1.2 * y

solution = solve_ivp(
    growth,
    [0, 15],
    [20],
    dense_output=True
)

t = np.linspace(0, 15, 200)

plt.figure(figsize=(8,5))
plt.plot(t, solution.sol(t)[0])

plt.title("Exponential Growth")
plt.xlabel("Time")
plt.ylabel("y")
plt.grid()
plt.show()
```

### Problem 2

Solve

$$

$$

with

$$

Plot the solution.



```python
def growth(t, y):
    return 3 * y

solution = solve_ivp(
    growth,
    [0, 3],      # time interval
    [2],         # initial condition y(0)=2
    dense_output=True
)

t = np.linspace(0, 3, 200)

plt.figure(figsize=(8,5))
plt.plot(t, solution.sol(t)[0])

plt.title("Solution of dy/dt = 3y")
plt.xlabel("t")
plt.ylabel("y")
plt.grid()

plt.show()
```

## Problem 4

Modify the logistic model by ranging $K$ from 0 to 50 with 100 slices.

How does the solution change?


```python
r = 0.5
y0 = 5
t = np.linspace(0, 20, 200)

plt.figure(figsize=(8,5))

for K in np.linspace(0.5, 50, 100):
    def logistic(t, y):
        return r * y * (1 - y / K)

    solution = solve_ivp(
        logistic,
        [0, 20],
        [y0],
        t_eval=t
    )

    plt.plot(solution.t, solution.y[0], alpha=0.2)

plt.title("Logistic Growth for Different Carrying Capacities")
plt.xlabel("Time")
plt.ylabel("Population")
plt.grid()
```

## Problem 5

Create your own model.

Ideas include

- Predator-prey model
- Cooling of coffee
- Disease spread
- Population growth
- Bank account with continuous interest

Model the system using a differential equation and solve it with `solve_ivp()`

```python
T_env = 22

def cooling(t, T):
    return -0.2 * (T - T_env)

solution = solve_ivp(
    cooling,
    [0, 30],
    [90],
    dense_output=True
)

t = np.linspace(0, 30, 200)

plt.figure(figsize=(8,5))
plt.plot(t, solution.sol(t)[0])

plt.title("Cooling of Coffee")
plt.xlabel("Time (minutes)")
plt.ylabel("Temperature (C)")
plt.grid()

plt.show()
```
