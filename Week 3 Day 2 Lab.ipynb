### Problem 1

Plot the function $f(x)=x^2$ for $x\in [-1,1]$ using `np.linspace()` with 10, 100, 1000 slices. Make sure your plot is as informative as possible.

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(-1,1,10)
x2 = np.linspace(-1,1,100)
x3 = np.linspace(-1,1,1000)

y1 = x1**2
y2 = x2**2
y3 = x3**2

plt.plot(
    x1, y1,
    label='10 sample points'
)

plt.plot(
    x2, y2,
    label='100 sample points'
)

plt.plot(
    x3, y3,
    label='1000 sample points'
)

plt.title(
    'Graph of the Quadratic Function $f(x)=x^2$\n'
    'Comparison of Different Sampling Resolutions',
    fontsize=16,
    fontweight='bold'
)

plt.xlabel(
    'Input Variable, x',
    fontsize=13
)

plt.ylabel(
    'Function Value, $f(x)=x^2$',
    fontsize=13
)

plt.legend(
    title='Sampling Resolution',
    fontsize=11,
    title_fontsize=12,
    loc='upper center'
)
```

### Problem 4
Download the "WA_County_Boundaries.shp" file. 
- Plot the state using `name_of_file_read.plot()`
- Plot some information about the state by extracting two columns of data.

```python
import geopandas as gpd
import matplotlib.pyplot as plt

wa = gpd.read_file("WA_County_Boundaries.shp")

fig, ax = plt.subplots(figsize=(8, 10))

wa.plot(
    ax=ax,
    color="lightgreen",
    edgecolor="white",
    linewidth=0.8
)

ax.set_title("Washington Counties", fontsize=18, fontweight="bold", pad=15)
ax.set_axis_off()
fig.patch.set_facecolor("white")
```
