### Problem 3
Create one dataset with a strong positive correlation and another with a strong negative correlation. Plot them side by side and compare their correlation coefficients.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

x = np.arange(1, 21)

y_positive = 2 * x + np.random.normal(0, 2, len(x))

y_negative = -2 * x + np.random.normal(0, 2, len(x))

corr_positive = np.corrcoef(x, y_positive)[0, 1]
corr_negative = np.corrcoef(x, y_negative)[0, 1]

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(x, y_positive, color='blue')
plt.title(f'Positive Correlation\nr = {corr_positive:.3f}')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(1, 2, 2)
plt.scatter(x, y_negative, color='red')
plt.title(f'Negative Correlation\nr = {corr_negative:.3f}')
plt.xlabel('x')
plt.ylabel('y')


print("Positive correlation coefficient", corr_positive)
print("Negative correlation coefficient", corr_negative)
```
